````python
# Import all NLP-Hub modules
import sys
sys.path.append('..')

from nlp_hub import (
    TextPreprocessor, Tokenizer,
    SentimentAnalyzer,
    TextClassifier, TopicModeler,
    TransformerModels,
    ChatBot, ConversationManager,
    NLPVisualizer
)

import pandas as pd
import matplotlib.pyplot as plt
````

## 1. Text Preprocessing & Tokenization

````python
# Sample texts
texts = [
    "Natural Language Processing (NLP) is fascinating! 🚀",
    "Machine Learning algorithms help computers understand human language.",
    "The future of AI lies in better language understanding."
]

# Initialize preprocessor
preprocessor = TextPreprocessor()
tokenizer = Tokenizer()

for i, text in enumerate(texts):
    print(f"\n=== Text {i+1} ===")
    print(f"Original: {text}")
    
    # Basic preprocessing
    cleaned = preprocessor.clean_text(text)
    tokens = preprocessor.preprocess_text(text)
    
    print(f"Cleaned: {cleaned}")
    print(f"Tokens: {tokens}")
    
    # Advanced tokenization
    advanced = tokenizer.advanced_tokenize(text)
    print(f"POS Tags: {advanced.get('pos_tags', [])}")
    print(f"Entities: {advanced.get('named_entities', [])}")
````

## 2. Sentiment Analysis

````python
# Sample texts with different sentiments
sentiment_texts = [
    "I absolutely love this new technology! It's revolutionary!",
    "This product is terrible. Complete waste of money.",
    "It's an okay solution. Nothing spectacular but works fine.",
    "Outstanding performance! Highly recommend to everyone.",
    "Disappointing results. Expected much better quality."
]

# Initialize sentiment analyzer
analyzer = SentimentAnalyzer()

print("Sentiment Analysis Results:")
print("=" * 50)

results = []
for text in sentiment_texts:
    # Use ensemble method
    result = analyzer.ensemble_sentiment(text, methods=['textblob', 'vader'])
    
    results.append({
        'text': text[:40] + '...',
        'sentiment': result['ensemble_sentiment'],
        'confidence': result['ensemble_confidence']
    })
    
    print(f"Text: {text}")
    print(f"Sentiment: {result['ensemble_sentiment']} (confidence: {result['ensemble_confidence']:.2f})")
    print()

# Create DataFrame
df_sentiment = pd.DataFrame(results)
print(df_sentiment)
````

## 3. Text Classification

````python
# Sample data for text classification
training_texts = [
    "Machine learning is a subset of artificial intelligence",
    "Deep learning uses neural networks with multiple layers",
    "The stock market reached new highs today",
    "Financial analysts predict strong quarterly results",
    "The latest smartphone features advanced camera technology",
    "New software update improves user interface design"
]

training_labels = [
    "technology", "technology", "finance", "finance", "technology", "technology"
]

# Initialize classifier
classifier = TextClassifier()

# Train the model
print("Training text classifier...")
training_results = classifier.train_classifier(
    training_texts, training_labels, model_name='logistic_regression'
)

print(f"Training accuracy: {training_results['accuracy']:.3f}")
print(f"Cross-validation score: {training_results['cv_mean']:.3f} ± {training_results['cv_std']:.3f}")

# Test on new data
test_texts = [
    "Artificial intelligence revolutionizes computing",
    "Investment portfolio shows positive returns"
]

predictions = classifier.predict(test_texts)
print("\nPredictions:")
for pred in predictions:
    print(f"Text: {pred['text']}")
    print(f"Prediction: {pred['prediction']} (confidence: {pred['confidence']:.3f})")
    print()
````

## 4. Topic Modeling

````python
# Sample documents for topic modeling
documents = [
    "Machine learning algorithms process data to find patterns",
    "Neural networks mimic brain functionality for computing",
    "Stock market analysis helps investors make decisions",
    "Financial planning requires careful budget management",
    "Climate change affects global weather patterns",
    "Environmental protection is crucial for sustainability",
    "Artificial intelligence transforms business operations",
    "Deep learning advances computer vision capabilities"
]

# Initialize topic modeler
topic_modeler = TopicModeler()

# Train LDA model
print("Training topic model...")
lda_results = topic_modeler.train_lda_sklearn(documents, num_topics=3)

print("\nDiscovered Topics:")
for topic in lda_results['topics']:
    print(f"\nTopic {topic['topic_id'] + 1}:")
    top_words = topic['words'][:5]
    print(f"Top words: {', '.join(top_words)}")

print(f"\nModel perplexity: {lda_results['perplexity']:.2f}")
````

## 5. Chatbot Interaction

````python
# Initialize chatbot
bot = ChatBot("NLP-Hub Assistant")

# Sample conversation
messages = [
    "Hello! How are you today?",
    "What can you help me with?",
    "Can you analyze the sentiment of this text: I'm feeling great!",
    "Tell me about natural language processing",
    "Thank you for your help!"
]

print("Chatbot Conversation:")
print("=" * 50)

for message in messages:
    response = bot.chat(message)
    
    print(f"User: {message}")
    print(f"Bot: {response['response']}")
    print(f"Intent: {response['analysis']['intent']}")
    print(f"Sentiment: {response['analysis']['sentiment']['sentiment']}")
    print()

# Get conversation statistics
stats = bot.get_stats()
print(f"Conversation Statistics:")
print(f"Total messages: {stats['total_messages']}")
print(f"Intent distribution: {stats['intent_distribution']}")
print(f"Sentiment distribution: {stats['sentiment_distribution']}")
````

## 6. Data Visualization

````python
# Initialize visualizer
visualizer = NLPVisualizer()

# Sample data for visualization
sample_texts = [
    "artificial intelligence machine learning deep learning neural networks",
    "natural language processing text analysis sentiment classification",
    "data science analytics visualization insights patterns discovery"
]

sample_sentiments = ["positive", "negative", "neutral", "positive", "negative"]

print("Creating visualizations...")

# 1. Word Cloud
fig1 = visualizer.word_cloud(
    " ".join(sample_texts), 
    width=600, 
    height=300,
    colormap='viridis'
)
plt.figure(figsize=(10, 5))
plt.title("NLP Topics Word Cloud")
plt.show()

# 2. Sentiment Distribution
fig2 = visualizer.sentiment_distribution(
    sample_sentiments, 
    "Sample Sentiment Distribution"
)
plt.show()

# 3. Text Length Distribution
fig3 = visualizer.text_length_distribution(
    sample_texts + sentiment_texts,
    "Text Length Analysis"
)
plt.show()

print("Visualizations completed!")
````

## 7. Transformer Models (Optional)

````python
# Note: This section requires transformers library and may take time to download models

try:
    # Initialize transformer models
    transformer_models = TransformerModels()
    
    # Load sentiment analysis pipeline
    success = transformer_models.load_pipeline('sentiment-analysis')
    
    if success:
        # Test transformer sentiment analysis
        test_texts = [
            "This is absolutely amazing!",
            "I don't like this at all.",
            "It's okay, nothing special."
        ]
        
        results = transformer_models.classify_text(test_texts)
        
        print("Transformer Sentiment Analysis:")
        for result in results:
            if 'error' not in result:
                print(f"Text: {result['text']}")
                print(f"Sentiment: {result['label']} (confidence: {result['confidence']:.3f})")
                print()
    else:
        print("Could not load transformer models. Install transformers library.")
        
except ImportError:
    print("Transformers library not available. Install with: pip install transformers torch")
````
