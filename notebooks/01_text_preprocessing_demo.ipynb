```python
import sys
sys.path.append('..')

from nlp_hub.preprocessing import TextPreprocessor, Tokenizer
import pandas as pd
```

## Sample Text Data

```python
sample_texts = [
    "Hello! This is a GREAT example of text preprocessing. It contains punctuation, UPPERCASE, and numbers like 123.",
    "Natural Language Processing (NLP) is amazing! It helps computers understand human language.",
    "The quick brown fox jumps over the lazy dog. This sentence contains every letter of the alphabet!",
    "Machine Learning and AI are revolutionizing how we process and analyze textual data."
]
```

## Text Preprocessing

```python
# Initialize preprocessor
preprocessor = TextPreprocessor()

# Process each text
for i, text in enumerate(sample_texts):
    print(f"\n=== Text {i+1} ===")
    print(f"Original: {text}")
    
    # Basic cleaning
    cleaned = preprocessor.clean_text(text)
    print(f"Cleaned: {cleaned}")
    
    # Full preprocessing
    processed = preprocessor.preprocess_text(text)
    print(f"Processed tokens: {processed}")
```

## Advanced Tokenization

```python
# Initialize tokenizer
tokenizer = Tokenizer()

# Advanced tokenization on first text
text = sample_texts[0]
print(f"Text: {text}\n")

# Basic tokenization
words = tokenizer.word_tokenize(text)
print(f"Word tokens: {words}\n")

# Sentence tokenization
sentences = tokenizer.sentence_tokenize(text)
print(f"Sentences: {sentences}\n")

# Advanced analysis
advanced = tokenizer.advanced_tokenize(text)
print(f"Advanced analysis:")
for key, value in advanced.items():
    print(f"  {key}: {value}")
```

## N-gram Extraction

```python
# Extract n-grams
text = "Natural Language Processing is a fascinating field of study"
tokens = tokenizer.word_tokenize(text.lower())

print(f"Text: {text}")
print(f"Tokens: {tokens}\n")

# Bigrams
bigrams = tokenizer.extract_ngrams(tokens, 2)
print(f"Bigrams: {bigrams}\n")

# Trigrams
trigrams = tokenizer.extract_ngrams(tokens, 3)
print(f"Trigrams: {trigrams}")
```

## Phrase Extraction

```python
# Extract noun phrases
text = "The artificial intelligence research team at the university developed an innovative machine learning algorithm."
phrases = tokenizer.extract_phrases(text)

print(f"Text: {text}")
print(f"Noun phrases: {phrases}")
```

## Batch Processing

```python
# Process multiple texts
results = []

for text in sample_texts:
    processed = preprocessor.preprocess_text(
        text, 
        remove_stopwords=True,
        apply_lemmatization=True
    )
    
    results.append({
        'original': text,
        'processed': processed,
        'token_count': len(processed)
    })

# Create DataFrame for better visualization
df = pd.DataFrame(results)
print(df)
```
