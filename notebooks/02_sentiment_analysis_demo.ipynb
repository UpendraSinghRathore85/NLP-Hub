# NLP-Hub: Sentiment Analysis Demo

This notebook demonstrates the sentiment analysis capabilities using multiple approaches.

```python
import sys
sys.path.append('..')

from nlp_hub.sentiment import SentimentAnalyzer
from nlp_hub.visualization import NLPVisualizer
import pandas as pd
import matplotlib.pyplot as plt
```

## Sample Data

```python
sample_texts = [
    "I absolutely love this product! It's amazing and works perfectly.",
    "This is terrible. I hate it and want my money back.",
    "The weather today is okay, nothing special.",
    "Outstanding performance! Highly recommend to everyone.",
    "Worst experience ever. Complete waste of time and money.",
    "It's an average product. Neither good nor bad.",
    "Fantastic! Best purchase I've made this year.",
    "Disappointing quality. Expected much better.",
    "Pretty good overall. Some minor issues but mostly satisfied.",
    "Excellent customer service and fast delivery!"
]
```

## Initialize Sentiment Analyzer

```python
# Initialize sentiment analyzer and visualizer
analyzer = SentimentAnalyzer()
visualizer = NLPVisualizer()
```

## Single Text Analysis

```python
text = sample_texts[0]
print(f"Text: {text}\n")

# TextBlob analysis
textblob_result = analyzer.textblob_sentiment(text)
print(f"TextBlob: {textblob_result}\n")

# VADER analysis
vader_result = analyzer.vader_sentiment(text)
print(f"VADER: {vader_result}\n")

# Transformer analysis (if available)
transformer_result = analyzer.transformer_sentiment(text)
print(f"Transformer: {transformer_result}\n")

# Ensemble analysis
ensemble_result = analyzer.ensemble_sentiment(text)
print(f"Ensemble: {ensemble_result}")
```

## Batch Analysis

```python
# Analyze all texts using ensemble method
results = analyzer.batch_analyze(sample_texts, method='ensemble')

# Create DataFrame for better visualization
data = []
for i, (text, result) in enumerate(zip(sample_texts, results)):
    data.append({
        'text': text[:50] + '...' if len(text) > 50 else text,
        'sentiment': result.get('ensemble_sentiment', 'unknown'),
        'confidence': result.get('ensemble_confidence', 0)
    })

df = pd.DataFrame(data)
print(df)
```

## Sentiment Distribution Visualization

```python
# Extract sentiments for visualization
sentiments = [result.get('ensemble_sentiment', 'unknown') for result in results]

# Create sentiment distribution plot
fig = visualizer.sentiment_distribution(sentiments, "Sample Texts Sentiment Distribution")
plt.show()
```

## Emotion Analysis

```python
# Analyze emotions for a few sample texts
emotion_texts = sample_texts[:3]

for i, text in enumerate(emotion_texts):
    emotions = analyzer.analyze_text_emotions(text)
    print(f"\nText {i+1}: {text}")
    print("Emotions:")
    for emotion, score in emotions.items():
        print(f"  {emotion}: {score:.3f}")
```

## Method Comparison

```python
# Compare different methods on the same texts
methods = ['textblob', 'vader']
comparison_results = {}

for method in methods:
    method_results = analyzer.batch_analyze(sample_texts[:5], method=method)
    comparison_results[method] = [r.get('sentiment', 'unknown') for r in method_results]

# Create comparison DataFrame
comparison_df = pd.DataFrame(comparison_results)
comparison_df['text'] = [text[:40] + '...' for text in sample_texts[:5]]
comparison_df = comparison_df[['text'] + methods]

print("Method Comparison:")
print(comparison_df)
```

## Interactive Sentiment Radar Chart

```python
# Create sentiment scores for radar chart
text = "I love this amazing product! It's fantastic and brings me so much joy."
emotions = analyzer.analyze_text_emotions(text)

# Create interactive radar chart
fig = visualizer.sentiment_radar_chart(emotions, f"Emotion Analysis: {text[:30]}...")
fig.show()
```

## Training Custom ML Model

```python
# Create simple training data
training_texts = [
    "I love this product", "Amazing quality", "Fantastic experience",
    "Terrible service", "Hate this item", "Worst purchase ever",
    "It's okay", "Average product", "Nothing special"
]

training_labels = [
    "positive", "positive", "positive",
    "negative", "negative", "negative",
    "neutral", "neutral", "neutral"
]

# Train ML model
print("Training custom ML model...")
training_results = analyzer.train_ml_model(training_texts, training_labels)
print(f"Training Results: {training_results}")

# Test on new data
test_text = "This product is absolutely wonderful!"
ml_result = analyzer.ml_sentiment(test_text)
print(f"\nML Prediction for '{test_text}': {ml_result}")
```
